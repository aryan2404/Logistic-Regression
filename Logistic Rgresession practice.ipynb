```python
plt.style.use('ggplot')
sns.scatterplot(x = 'total_bill', y ='tip', data = tips)
plt.show()
```

```python
sns.histplot(tips['total_bill'],bins = 10)
plt.show()
```

```python
#basis box plot
sns.boxplot(x = 'day', y ='total_bill', data = tips)
plt.show()
```

```python
Q1 = tips['total_bill'].quantile(0.25)
Q3 = tips['total_bill'].quantile(0.75)
IQR = Q3 -Q1
print(IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print()
print(upper_bound)
```

```python
filter1 = tips['total_bill']>lower_bound
filter2 = tips['total_bill']<upper_bound
tips_filtered = tips[filter1 & filter2]
```

```python
sns.boxplot(x = 'day',y ='total_bill', data = tips_filtered)
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'day', palette = 'coolwarm')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Feature1': [10,20,30,40,50],
    'Feature2': [100,150,200,250,300]
    
}
df = pd.DataFrame(data)
df
```

```python
standardized_df = (df-df.mean())/df.std()
print(standardized_df)
```

```python
df.sum()
```

```python
normalized_df = (df-df.mean()) / (df.max()-df.min())
print (normalized_df)
```
